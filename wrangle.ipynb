```python
# let's look at each column
for col in df.columns:
    print(f'-------- {col} column value_counts --------')
    print(f'-------- {df[col].nunique()} unique values -------')
    print(df[col].value_counts(sort=True))
    
```

```python
# facilityid looks like it needs to be an int. Check for nonzero after decimal point
for x in df.facilityid:
    if ((x % 1) != 0): print(df.facilityid)
```

```python
# all facilityid values are xxx.0, so make this an int
df.loc[:, 'facilityid'] = df.facilityid.astype(int)
```
